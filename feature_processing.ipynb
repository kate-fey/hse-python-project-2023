```python
def calculate_sentiment(text):
    # Run VADER on the text
    scores = sentimentAnalyser.polarity_scores(text)
    # Extract the compound score
    compound_score = scores['compound']
    # Return compound score
    return compound_score
```

```python
def psych_capital(text): # функция, которая добавляет балл в психологический капитал за упоминание опр. слов
    score = 0
    for word in text:
        if word in optimism_list:
            score += 1
        if word in hope_list:
            score += 1
        if word in resilience_list:
            score += 1
        if word in confidence_list:
            score += 1
    return score
```

```python
def innovative(text):  # функция, которая добавляет балл в инновационность за упоминание опр. слов
    score = 0
    for word in text:
        if word in innovation_list:
            score += 1
    return score
```

```python
def eco(text):  #функция, которая добавляет балл в экологичность за упоминание опр. слов
    score = 0
    for word in text:
        if word in environment_list:
            score += 1
    return score
```

```python
# функция, которая собирает суп страницы
from selenium import webdriver
from bs4 import BeautifulSoup

def get_soup(url):
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get(url)
    pageSource = driver.page_source
    driver.quit()
    return BeautifulSoup(pageSource , "html.parser")
```

```python
soup = get_soup('https://www.city-town.uk/country-england.html?page=2')
```

```python
# собираем имена всех городов со страницы
def get_cit(soup):
    pagedata = soup.find_all('h2', {'class': 'alv'})
    cit_list = []
    for item in pagedata:
        town = item.a.get('title')
        cit_list.append(town)
    return cit_list
```

```python
get_cit(soup)
```

```python
# проходим по всем страницам
from tqdm.notebook import tqdm
import time

MAIN_URL = 'https://www.city-town.uk/country-england.html?'

data = []
fails = []
for p in tqdm(range(1, 720)):
    url = MAIN_URL + f'page={p}'
    try:
        tree = get_soup(url)
        cities = get_cit(tree)
        data.extend(cities)
        time.sleep(0.1)
    except:
        fails.append(url)
```
